```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
vocab_size= 10000 
```

# One Hot Representation

```python
onehot_repn= [one_hot(words,vocab_size) for words in sent]
onehot_repn
```

# Words Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# To make every sentence of equal length.
sent_length=8
embedded_docs=pad_sequences(onehot_repn,padding='pre',maxlen=sent_length)   # PRE- Adds 0 in first.
print(embedded_docs)
```

```python
dim= 10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
